# HEXACO-PI-R 100 Survey Results: Agents 'v' Humans

---

### Results from Llama-3.2 Agents:

```python
import pandas as pd
from support.hexaco_pi_r import scoring, responses_to_scores
model_label = 'llama3.2'
responses_df = pd.read_csv(f'data/hexaco-pi-r-responses-{model_label}.csv', index_col=0)
responses_df.columns = range(len(responses_df.columns))
scores_df = responses_to_scores(responses_df, scoring)
scores_df[:5]
```

```python
from support.hexaco_pi_r import calc_dimension_scores
dimensions_df = calc_dimension_scores(scores_df, scoring)
dimensions_df[:5]
```

```python
from support.hexaco_pi_r import calc_mean_sd_alpha
df_stats = calc_mean_sd_alpha(scores_df, scoring, own_alpha=False)
pd.options.display.float_format = "{:,.4f}".format
df_stats
```

```python
from support.hexaco_pi_r import plot
plot(df_stats, f'Agents({model_label}) vs Human HEXACO-PI-R 100 Scores')
```
